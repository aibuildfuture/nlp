```python
import os
from pyltp import Postagger
from pyltp import Segmentor
from pyltp import NamedEntityRecognizer
from pyltp import Parser
LTP_DATA_DIR = 'ltp_data_v3.4.0/'  # ltp模型目录的路径

cws_model_path = os.path.join(LTP_DATA_DIR, 'cws.model')  # 分词模型路径，模型名称为`cws.model`
segmentor = Segmentor()  # 初始化实例
segmentor.load(cws_model_path)  # 加载模型

pos_model_path = os.path.join(LTP_DATA_DIR, 'pos.model')  # 词性标注模型路径，模型名称为`pos.model`
postagger = Postagger() # 初始化实例
postagger.load(pos_model_path)  # 加载模型

ner_model_path = os.path.join(LTP_DATA_DIR, 'ner.model')  # 命名实体识别模型路径，模型名称为`pos.model`
recognizer = NamedEntityRecognizer() # 初始化实例
recognizer.load(ner_model_path)  # 加载模型

par_model_path = os.path.join(LTP_DATA_DIR, 'parser.model')  # 依存句法分析模型路径，模型名称为`parser.model`
parser = Parser() # 初始化实例
parser.load(par_model_path)  # 加载模型
```

```python
def get_keywords(article):
    keywords = []#关键词tuple列表
    for word_similar_to_say in words_similar_to_say:
        start_index = 0
        index = article.find(word_similar_to_say, start_index)
        while(index > -1):
            keywords.append((word_similar_to_say,index))
            start_index = index + 1
            index = article.find(word_similar_to_say, start_index)
    return keywords
```

```python
def get_statments(keywords):
    statments = []#关键词附近的句子
    for keyword_tuple in keywords:
        keyword = keyword_tuple[0]
        keyword_index = keyword_tuple[1]
        #前一个句号的index
        pre_stop_index = article.rfind('。',0,keyword_index)
        #后一个句号的index
        post_stop_index = article.find('。',keyword_index)
        statment = article[pre_stop_index:post_stop_index]
        statments.append(statment)
    return statments
```

```python
def get_root_index(arcs):#根据依存分析结果获取根索引
    root_index = 0
    for i,arc in enumerate(arcs):
        if(arc.head == 0):
            root_index = i
    return root_index
```

```python
def get_subject_index(arcs, root_index):#根据依存分析结果和根索引获取主语索引
    subject_index = 0
    for i,arc in enumerate(arcs):
        if(arc.head == root_index + 1 and arc.relation == 'SBV'):
            subject_index = i
    return subject_index
```

```python
def get_object_index(arcs,root_index):#根据依存分析结果和根索引获取宾语索引
    object_index = 0
    for i,arc in enumerate(arcs):
        if(arc.head == root_index + 1 and arc.relation == 'VOB'):
            object_index = i
    return object_index
```

```python
def get_content(object_index, subject_index, words):#根据主语宾语，从分词列表中获取言论内容
    speech_content = ''
    if(object_index > subject_index):
        start_index = root_index
        speech_content = ''.join(words[root_index+1:len(words)])
    return speech_content
```

```python
def get_speeches(statments):
    #对每个包含关键词的句子，提取言论
    speeches = []
    for i,statment in enumerate(statments):
        words = segmentor.segment(statment)  # 分词
        #print ('\t'.join(words))
        postags = postagger.postag(words)  # 词性标注
        #print('\t'.join(postags))
        #netags = recognizer.recognize(words, postags)  # 命名实体识别
        arcs = parser.parse(words, postags)  # 句法分析
        #print ("\t".join("%d:%s" % (arc.head, arc.relation) for arc in arcs))
        if(keywords[i][0] in words):#关键词在分词列表中
            root_index = get_root_index(arcs)
            subject_index = get_subject_index(arcs, root_index)
            subject = words[subject_index]
            object_index = get_object_index(arcs, root_index)
            content = get_content(object_index, subject_index, words)
            speeches.append((subject, content))
    return speeches
```
