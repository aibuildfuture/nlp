CNN由Yan LeCun在1988年提出，首次实现的网络结构是LeNet，用于手写数字识别问题。
LeNet-5 参数: 60k
网络结构: Conv -> Pool -> Conv -> Pool -> FC -> FC -> Output

AlexNet是Alex在2012年提出，在ImageNet竞赛里，把top5分类错误率由前一年的25%降低到16%。
相对于LeNet，增加了更多的卷积层和池化层。激活函数由tahh变为ReLu。参数达到6千万。

VGG-16在2014年提出，增加了更多层，层数达到16层。  
卷积层使用3\*3的filter，值是1的stride，池化层使用2\*2的filter和值是2的stride。



GoogLeNet是2015年提出的。用多个filter做卷积操作，然后把结果拉平作为输出。这样可以提取更多信息。

ResNet是微软亚洲研究院在2015年提出的。深度达到上千层，识别率超过人类。添加了残差信息，解决了训练误差增加问题。随着网络加深，输入数据被多次处理，特征可能不再明显，导致训练误差增加。ResNet计算激活值时，把两层前的激活值作为残差和本层的线性函数的输出的和作为本层激活函数的输入。这样保留了前面层的更多信息。减小了训练误差。
